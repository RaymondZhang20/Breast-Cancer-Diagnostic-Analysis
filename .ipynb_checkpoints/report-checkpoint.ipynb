## Imports 

```python
import os
import sys
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.compose import ColumnTransformer, make_column_transformer
from sklearn.dummy import DummyClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    f1_score,
    make_scorer,
    ConfusionMatrixDisplay,
)
from sklearn.model_selection import (
    GridSearchCV,
    RandomizedSearchCV,
    cross_val_score,
    cross_validate,
    train_test_split,
)
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.svm import SVC
```

## 1. Introduction <a name="in"></a>
<hr>

Breast cancer was one of the most often described forms of cancer in ancient documents because the necrotic tumor could be felt through the skin. Even now, breast cancer is the most commonly diagnosed invasive cancer besides lung cancer. In this report, I want to discover which features of breast tumors tend to relate to their malignancy, how should be diagnosed as malignant and how should we classify them. For this problem, I will use Breast Cancer Wisconsin (Diagnostic) Data Set on Kaggle(the data set is available [here](https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data?resource=download) or [here](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Diagnostic%29)). 

In this data set, there are 426 examples and 32 columns, and the goal is: 
1. to predict whether breast cancer is malignant and explore the feature's importance. 
2. to separate breast cancer into groups.

```python
data = pd.read_csv("data.csv", encoding="utf-8")
```

So in feature engineering, these six columns and three columns may be integrated in two columns.

## 4. Feature engineering <a name="2"></a>
<hr>

Perform feature reduction and feature selection and extract new features relevant for the problem.

```python
train_df = train_df.iloc[:, 1:32]
test_df = test_df.iloc[:, 1:32]
train_df.replace(('M', 'B'), (1, 0), inplace=True)
test_df.replace(('M', 'B'), (1, 0), inplace=True)

train_df['engineered_feature1'] = train_df.apply(lambda row: (row.radius_mean + row.perimeter_mean + row.area_mean + row.radius_worst + row.perimeter_worst + row.area_worst)/6, axis=1)
train_df['engineered_feature2'] = train_df.apply(lambda row: (row.radius_se + row.perimeter_se + row.area_se)/3, axis=1)
test_df['engineered_feature1'] = train_df.apply(lambda row: (row.radius_mean + row.perimeter_mean + row.area_mean + row.radius_worst + row.perimeter_worst + row.area_worst)/6, axis=1)
test_df['engineered_feature2'] = train_df.apply(lambda row: (row.radius_se + row.perimeter_se + row.area_se)/3, axis=1)
```

## 5. Preprocessing and Transformations <a name="2"></a>
<hr>

Identify different feature types and the transformations I would apply on each feature type and define a column transformer.

```python
X_train = train_df.drop(columns=["diagnosis"])
X_test = test_df.drop(columns=["diagnosis"])

y_train = train_df["diagnosis"]
y_test = test_df["diagnosis"]

numeric_features = ["engineered_feature1", "engineered_feature2", "texture_mean", "smoothness_mean", "compactness_mean", "concavity_mean",
                    "concave points_mean", "symmetry_mean", "fractal_dimension_mean", "texture_se", "smoothness_se", "compactness_se",
                    "concavity_se", "concave points_se", "symmetry_se", "fractal_dimension_se", "texture_worst", "smoothness_worst", "compactness_worst",
                    "concavity_worst", "concave points_worst", "symmetry_worst", "fractal_dimension_worst"]
binary_features = []
ordinal_features = []
drop_features = ["radius_mean", "perimeter_mean", "area_mean", "radius_worst", "perimeter_worst", "area_worst", "radius_se", "perimeter_se", "area_se"]
preprocessor  = make_column_transformer(
    (StandardScaler(), numeric_features),
    ("passthrough", binary_features),
    (SimpleImputer(missing_values=np.nan, strategy='mean'), ordinal_features),
    ("drop", drop_features),)
```

## 6. Baseline model <a name="6"></a>
<hr>

Conduct supervised learning by dummyclassifier.

```python
baseline = DummyClassifier()
pipe = make_pipeline(preprocessor, baseline)
scores = cross_validate(
    pipe, X_train, y_train, return_train_score=True
)
pd.DataFrame(scores)
```

## 7. Other models <a name="7"></a>
<hr>

1. Try logistic regression, SVC, LightGBM,and XGBoost . 
2. Carry out hyperparameter tuning to explore different values for the complexity hyperparameter `C`. 
3. Report validation scores along with standard deviation.

#### Define a function that return mean validation scores with deviation:

```python
def mean_std_cross_val_scores(model, X_train, y_train, **kwargs):
    scores = cross_validate(model, X_train, y_train, **kwargs)
    mean_scores = pd.DataFrame(scores).mean()
    std_scores = pd.DataFrame(scores).std()
    out_col = []

    for i in range(len(mean_scores)):
        out_col.append((f"%0.3f (+/- %0.3f)" % (mean_scores[i], std_scores[i])))

    return pd.Series(data=out_col, index=mean_scores.index)
```

#### Finding the best C for Logistic Regression:

```python
X_train_enc = pd.DataFrame(
    preprocessor.fit_transform(X_train), index=X_train.index, columns=numeric_features
)
log = LogisticRegression(max_iter = 1000)
param_grid = {
    "C": [0.001, 0.01, 0.1, 1.0, 10, 100],
}
f1 = make_scorer(f1_score , average='macro')
grid_search = GridSearchCV(
    estimator=log, param_grid=param_grid, cv=5, return_train_score=True, scoring=f1
)
grid_search.fit(X_train_enc, y_train)
grid_search.best_params_
```

#### Report scores using logistic regression with best C:

```python
pipe = make_pipeline(preprocessor, LogisticRegression(C = 10, max_iter = 1000))
scores_mean_std = {}
scoring = [
    "accuracy",
    "f1",
    "recall",
    "precision",
]
scores_mean_std["logistic regression"] = mean_std_cross_val_scores(
    pipe, X_train, y_train, return_train_score=True, scoring=scoring
)
pd.DataFrame(scores_mean_std)
```

#### Finding the best C for SVC:

```python
X_train_enc = pd.DataFrame(
    preprocessor.fit_transform(X_train), index=X_train.index, columns=numeric_features
)
svc = SVC()
param_grid = {
    "C": [0.001, 0.01, 0.1, 1.0, 10, 100],
}
f1 = make_scorer(f1_score , average='macro')
grid_search = GridSearchCV(
    estimator=svc, param_grid=param_grid, cv=5, return_train_score=True, scoring=f1
)
grid_search.fit(X_train_enc, y_train)
grid_search.best_params_
```

#### Report scores using logistic regression with best C:

```python
pipe_svm = make_pipeline(preprocessor, SVC(C = 10))
scores_mean_std["SVC"] = mean_std_cross_val_scores(
        pipe_svm, X_train, y_train, return_train_score=True, scoring=scoring
    )
pd.DataFrame(scores_mean_std)
```

### Then, try two tree-based ensemble model XGBclassifier and LGBMClassifier.

```python
from xgboost import XGBClassifier
from lightgbm.sklearn import LGBMClassifier
pipe_lgbm = make_pipeline(preprocessor, LGBMClassifier(random_state=1))
scores_mean_std["LightGBM"] = mean_std_cross_val_scores(
        pipe_lgbm, X_train, y_train, return_train_score=True, scoring=scoring
    )
pipe_xgb = make_pipeline(
    preprocessor, XGBClassifier(random_state=1, eval_metric="logloss", verbosity=0)
)
scores_mean_std["XGBoost"] = mean_std_cross_val_scores(
        pipe_xgb, X_train, y_train, return_train_score=True, scoring=scoring
    )
pd.DataFrame(scores_mean_std)
```

### Summary:
We can see that all four model did a very good job at predicting the malignancy of breast cancer. Among four model, XGBoost have the highest test f1 score and test recall. All model have more than 0.9 on all scores so there is no overfitting or underfitting, and the standard deviations are all less than 0.05. So we can trust our model's result.
